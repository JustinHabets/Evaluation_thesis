## Process description

We use the BPIC 2017 event log, which records data about the process of loan applications from a Dutch financial institute. 

The log contains 31,509 cases, where each case represents a single loan application consisting of multiple events and associated attributes.

## Load Event Log

The cell below will load the BPI Challenge 2017 Event log, and prepares it for analysis with PM4Py.

```python
#Load event log

import pm4py
import pandas as pd
import numpy as np

df = pd.read_csv("BPIC17_eval.zip")
df = df.sort_values(["CaseID", "Timestamp"], ignore_index=True)
log = pm4py.format_dataframe(df, case_id='CaseID', activity_key='Activity', timestamp_key='Timestamp')

display(log)
```

```python
# Unique case count
print("\nNumber of cases:", log['CaseID'].nunique())

# Number of unique events (activities)
print("Number of unique activities:", log['Activity'].nunique())

print("\nOverview of event log attributes:")
print(log.dtypes)
```

### Subset filters

The distinction between the two groups is based on whether the activity *'A_Submitted'* appears in the case:
- website applications contain the activity *'A_Submitted'* 
- bank applications do **not** contain *'A_Submitted'*

The code below creates these subsets using *PM4Py*.

```python
log_website = pm4py.filter_event_attribute_values(log, "concept:name", ["A_Submitted"], level="case")
print("\nNumber of applications through website:", log_website['CaseID'].nunique())

log_bank = pm4py.filter_event_attribute_values(log, "concept:name", ["A_Submitted"], level="case", retain=False)
print("\nNumber of applications through bank:", log_bank['CaseID'].nunique())
```
